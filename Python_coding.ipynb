```python
# Load GDP data from CSV file
df = pd.read_csv('/content/GDP (nominal) per Capita.csv')
df.head()
```

## Data Cleaning

```python
# Display dataset info
print(f"Dataset loaded: GDP (nominal) per Capita.csv")
print(f"Shape: {df.shape}")
```

```python
# Remove unnecessary index column and strip whitespace from country names
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

df['Country/Territory'] = df['Country/Territory'].str.strip()
```

```python
# Check data types and missing values
print("Data Info:")
print(df.info())
```

```python
# Display missing values per column
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Handle zero values in IMF estimates (treat as missing data)
# Note: 0 values in IMF_Estimate and IMF_Year likely represent missing data
df.loc[df['IMF_Estimate'] == 0, 'IMF_Estimate'] = np.nan
df.loc[df['IMF_Year'] == 0, 'IMF_Year'] = np.nan

print("Updated missing values after handling zeros:")
print(df.isnull().sum())
```

## Exploratory Data Analysis

```python
# Summary statistics for numeric columns
print("Summary Statistics:")
print(df.describe())
```

```python
# Display first few rows
print("First 10 rows:")
df.head(10)
```

```python
# Count countries/territories by region
print("Countries/Territories by Region:")
print(df['UN_Region'].value_counts())
```

```python
# Check data availability across different sources
print("\nData Coverage by Source:")
print(f"IMF Estimates available: {(df['IMF_Estimate'].notna().sum() / len(df) * 100):.1f}%")
print(f"World Bank Estimates available: {(df['WorldBank_Estimate'].notna().sum() / len(df) * 100):.1f}%")
print(f"UN Estimates available: {(df['UN_Estimate'].notna().sum() / len(df) * 100):.1f}%")
```

## Data Visualization

```python
# Visualization: Distribution of UN estimates by region
plt.figure(figsize=(12, 6))
df.boxplot(column='UN_Estimate', by='UN_Region', figsize=(12, 6))
plt.title('GDP per Capita (UN Estimate) by Region')
plt.suptitle('')  # Suppress default title
plt.xlabel('Region')
plt.ylabel('GDP per Capita (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 15 countries by UN GDP estimate
top_15 = df.nlargest(15, 'UN_Estimate')[['Country/Territory', 'UN_Region', 'UN_Estimate', 'WorldBank_Estimate', 'IMF_Estimate']]
print("Top 15 Countries by GDP per Capita (UN Estimate):")
print(top_15)
```

```python
# Bottom 15 countries by UN GDP estimate (with available data)
bottom_15 = df.nsmallest(15, 'UN_Estimate')[['Country/Territory', 'UN_Region', 'UN_Estimate', 'WorldBank_Estimate', 'IMF_Estimate']]
print("Bottom 15 Countries by GDP per Capita (UN Estimate):")
print(bottom_15)
```
