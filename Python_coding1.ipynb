## Load the dataset
The cell below tries to read a known path (used when you uploaded the file to Colab) and falls back to an upload widget if the file is not found.
If you already mounted Google Drive and have the file there, you can change the path accordingly.

```python
from pathlib import Path
from google.colab import files

# Preferred path when you upload the CSV to Colab files area
default_paths = [
    '/content/GDP (nominal) per Capita.csv',
    '/content/student.csv'
]
data_path = None
for p in default_paths:
    if Path(p).exists():
        data_path = p
        break

if data_path is None:
    print("No default file found. Please upload the CSV file (example: 'GDP (nominal) per Capita.csv').")
    uploaded = files.upload()
    # take the first uploaded file
    if uploaded:
        data_path = '/content/' + next(iter(uploaded.keys()))

if data_path is None:
    raise FileNotFoundError("CSV file not found. Upload a file or adjust the path.")

# Read the CSV into a dataframe
df = pd.read_csv(data_path)

# Basic cleaning
if 'Country/Territory' in df.columns:
    df['Country/Territory'] = df['Country/Territory'].astype(str).str.strip()

print(f"Loaded: {data_path} — shape: {df.shape}")
df.head()
```

## Sample visualization — Top 10 by WorldBank estimate
This creates a horizontal bar chart of the top 10 countries by the WorldBank estimate column. If your column names differ, change the column used below.

```python
# Adjust column name here if needed
col = 'WorldBank_Estimate'
name_col = 'Country/Territory'

if col not in df.columns:
    raise KeyError(f"Column '{col}' not found in dataframe. Available columns: {list(df.columns)}")

# prepare data (drop missing / zero or negative values if appropriate)
plot_df = df[[name_col, col]].dropna()
# ensure numeric
plot_df[col] = pd.to_numeric(plot_df[col], errors='coerce')
plot_df = plot_df.dropna().sort_values(by=col, ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=plot_df, x=col, y=name_col, palette='viridis')
plt.title('Top 10 countries — WorldBank estimate (nominal GDP per capita)')
plt.xlabel(col)
plt.tight_layout()
plt.show()
```
